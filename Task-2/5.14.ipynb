```python
import bisect
import numpy as np
import math
import pylab
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from pylab import *
K = 0.016;
c_a0 = 42.0;
c_b0 = 28.0;
c_c0 = 4.0;
es  = 0.5;
def f(x):
    y = K - (c_c0 + x)/(((c_a0-2*x)**2) * (c_b0 - x));
    return y
x= np.linspace(0,20,200)
xl=0
xu=20
xr=(xu+xl)/2
Xr=[]
while abs(f(xr))>=0.0001:
    Xr.append(xr)
    if f(xl)*f(xr)>0:
        xl=xr
    else:
        xu=xr
    xr=(xu+xl)/2
    
print("the root of equation is =",xr)
plt.figure(figsize=(15,7 ))

FXR=[]
for j in Xr:
    FXR.append(f(j))
plt.scatter(Xr,FXR , marker='*',c='red',label='bisection method')
plt.plot(xr,f(xr),c='red',marker='o',label='root')
plt.plot(x,f(x),label='function plot')

plt.title('bisection')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.grid(True)

plt.show()

```
