```python
import numpy as np
import  matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
def analytical_bungee_jumper():
    t= np.arange(0 , 13 ,2)
    v =[]
    root = math.sqrt(9.81*68.1/0.25)
    root2 = math.sqrt(9.81*(0.25/68.1))
    for i in t :

        v.append(root*(np.tanh(root2*i)))
#   plt.plot(t,v ,label ='analytical')
    solution = pd.DataFrame(data = {"time(sec)":t , "velocity m/s" : v })
    return solution
```

```python

def approx_bungee_jumper(step):
    dt = step
    x= np.arange(0 , 12.5 , dt)
    y = []
    np.size(x)
    for i in range  (len(x)) :
        if (i == 0):
            y.append(0)
        else:
            y.append( y[i-1]+(9.81-(0.25/68.1)*(y[i-1])**2)*dt)     
    
    label = 'Step ='+ str(step)
#     plt.plot(x,y ,label=label )
    solution = pd.DataFrame(data = {"time(sec)":x , "velocity m/s" : y })
    return solution
```

```python

analytical_solution = analytical_bungee_jumper()
analytical_solution
```

```python
approx_05 =approx_bungee_jumper(0.5)
approx_05
```

```python
approx_1 =approx_bungee_jumper(1)
approx_1
```

```python
approx_2 =approx_bungee_jumper(2)
approx_2
```

```python
plt.subplot()
plt.grid()
plt.plot(analytical_solution['time(sec)'], analytical_solution['velocity m/s'] , label = 'analytical')
plt.plot(approx_05['time(sec)'], approx_05['velocity m/s'],label = 'step = 0.5')
plt.plot(approx_1['time(sec)'], approx_1['velocity m/s'], label = 'step = 1')
plt.plot(approx_2['time(sec)'], approx_2['velocity m/s'], label = 'step = 2' )
plt.legend()
```
