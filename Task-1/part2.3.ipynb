```python
import numpy as np
import  matplotlib.pyplot as plt
import math
import pandas as pd
```

```python

def approx_bungee_jumper(step):
#   for (a)
    dt = step
    x= np.arange(0 , 10.5 , dt)
    ya = []
    for i in range  (len(x)) :
        if (i == 0):
            ya.append(37)
        else:
            ya.append( ya[i-1]-0.12*dt*(ya[i-1]-10))

# fro (b)
    ta = -2*x +20
    yb = []
    for i in range  (len(x)) :
        if (i == 0):
            yb.append(37)
        else:
            yb.append( yb[i-1]-0.12*dt*(yb[i-1]-ta[i-1]))
    plt.grid()
    plt.subplot()
    plt.plot(x , ya , label = 'room-temp const')
    plt.plot(x , yb , label ='room-temp drops')
    plt.legend()
```

```python
approx_bungee_jumper(0.5)
```
